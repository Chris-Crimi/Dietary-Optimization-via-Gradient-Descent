```python
#removed cost
target_values = pd.read_excel(r"C:\Users\Chris\Desktop\Diet Data\Marisa Info.xlsx", sheet_name="Target Values", 
                              header = 0 #set column headers
                              )["Amounts"].to_list()
```

```python
from scipy.optimize import linprog

greater_than = ['Protein (g)','Total Fat (g)','Carbohydrate (g)','Fiber, total dietary (g)','Calcium (mg)','Folate, total (mcg)','Vitamin A, RAE (mcg_RAE)','Vitamin B-12 (mcg)','Vitamin B-6 (mg)','Vitamin C (mg)','Vitamin D (D2 + D3) (mcg)','Vitamin E (alpha-tocopherol) (mg)','Vitamin K (phylloquinone) (mcg)','Copper (mg)']
equal_to = ["Energy (kcal)"]

#need to figure out how to convert this to maximization subject to some slack variable
c = np.zeros_like(food_data["Energy (kcal)"].to_numpy())
A_ub = -food_data[greater_than].to_numpy().T
b_ub = -np.array(target_values[1:15])
A_eq = food_data[equal_to].to_numpy().T
b_eq = target_values[0]
bounds = [(req, limit) for req, limit in zip(food_data['Food Req'].to_numpy(),food_data['Food Limits'].to_numpy())]

results = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method="highs-ipm")

results
```

```python
results_frame = food_data[["Main food description"]].copy()
results_frame["Amount in 100g"] = results.x
results_frame
```

```python
nutrient_values_results = pd.DataFrame()
combined_constraints = equal_to + greater_than

additional_nutrients = ['Iron\n(mg)','Magnesium (mg)','Zinc\n(mg)','Phosphorus (mg)','Selenium (mcg)','Potassium (mg)','Sodium (mg)','Choline, total (mg)','Thiamin (mg)','Riboflavin (mg)','Niacin (mg)','18:3\n(g)']

nutrient_values_results["Nutrient"] = combined_constraints + additional_nutrients

nutrient_values_results["optimizer reults"] = np.dot(results.x, food_database[food_database["Food code"].isin(results_frame.index.to_list())][combined_constraints + additional_nutrients].to_numpy())
nutrient_values_results["target values"] = target_values[:]

nutrient_values_results
```

Nutrient Driver Assessment
- using gradient of nutrients wrt to each food to find the most important one for that nutrient
- steps are, we calculate the jacobian of the target values, assuming each nutrient is at 0. So to do this we multiply the diagonal of the target values by the food_data parameters we are opimizing along 

```python
d_foods = np.dot(np.diag(target_values[0:15]), food_data[combined_constraints].to_numpy().T)
driver_assessment = food_data[["Main food description"]].copy() # nutrient_values_results[["Nutrient"]].copy()
driver_assessment[combined_constraints] = d_foods.T
driver_assessment
```
